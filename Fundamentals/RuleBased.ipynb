**Rule based systems**

Rule-based systems are a type of artificial intelligence (AI) system that operates on the basis of a set of predefined rules. These rules, also known as "if-then" statements or production rules, dictate how the system should behave or make decisions based on the input it receives.

image.png

```python
def simple_chatbot(user_input):
    user_input = user_input.lower()
    if "hello" in user_input:
        return "Hello! How can I assist you today?"
    elif "how are you" in user_input:
        return "I'm just a program, but I'm functioning well. Thank you for asking!"
    else:
        return "I'm sorry, I didn't understand that. Can you please rephrase?"
```

```python

def app():
    print("Welcome to the Simple Chatbot!")
    print("Type 'bye' to exit.")

    while True:
        user_input = input("You: ")
        if user_input.lower() == 'bye':
            print("Chatbot: Goodbye! Have a great day!")
            break

        response = simple_chatbot(user_input)
        print("Chatbot:", response)

app()
```
